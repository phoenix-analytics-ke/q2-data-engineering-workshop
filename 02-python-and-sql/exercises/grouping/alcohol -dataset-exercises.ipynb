### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv). 

### Step 3. Assign it to a variable called drinks.

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv"
data=pd.read_csv(url)
drinks= pd.DataFrame(data)
drinks.head(10)
```

### Step 4. Which continent drinks more beer on average?

```python
#using groupby to find beer average
average_beer=drinks.groupby('continent')['beer_servings'].mean()
average_beer
```

### Step 5. For each continent print the statistics for wine consumption.

```python
#wine statistics
wine_stats=drinks.groupby('continent')['wine_servings'].describe()
wine_stats
```

### Step 6. Print the mean alcoohol consumption per continent for every column

```python
#alcohol consumption
avg_consumption=drinks.groupby('continent')[['beer_servings','spirit_servings','wine_servings']].mean()
avg_consumption
```

### Step 7. Print the median alcoohol consumption per continent for every column

```python
median_consumption=drinks.groupby('continent')[['beer_servings','spirit_servings','wine_servings']].median()
median_consumption
```

### Step 8. Print the mean, min and max values for spirit consumption.
#### This time output a DataFrame

```python

spirit_consumption=drinks.groupby('continent')['spirit_servings'].describe()
spirit_consumption
```
